```python
import pytrends
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
import matplotlib
import pandas as pd
```

# **Connect to Google Trends**

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

# **Make List of Keywords to Search**

```python
kw_list = ['face mask','hand sanitizer','fitness','recipes','toilet paper']
```

# **Choose the Date/Time Period and Location**

```python
pytrends.build_payload(kw_list, cat=0, timeframe='2020-01-01 2021-01-01', geo='US', gprop='')
```

# **Find the Interest Over Time**

```python
trends = pytrends.interest_over_time()
```

# **Drop Unnecessary Labels from Dataset**

```python
label_drop = trends.drop(labels=['isPartial'],axis='columns')
```

# **Configure Visualization**

```python
trend_visualization = label_drop.plot(title = 'COVID Trends 2020-2021')
```
